```python
import numpy as np
import time
import matplotlib.pyplot as plt

# Bubble sort
def bubble_sort(A):
    n = len(A)
    for i in range(n - 1, 0, -1):
        bChanged = False
        for j in range(i):
            if A[j] > A[j + 1]:
                A[j], A[j + 1] = A[j + 1], A[j]
                bChanged = True
        if not bChanged:
            break
    return A

# Insertion sort
def insertion_sort(A):
    n = len(A)
    for i in range(1, n):
        key = A[i]
        j = i - 1
        while j >= 0 and A[j] > key:
            A[j + 1] = A[j]
            j -= 1
        A[j + 1] = key
    return A

# Selection sort
def selection_sort(A):
    n = len(A)
    for i in range(n):
        least = i
        for j in range(i + 1, n):
            if A[least] > A[j]:
                least = j
        A[i], A[least] = A[least], A[i]
    return A

# Sequential search
def sequential_search(A, key, low, high):
    for i in range(low, high + 1):
        if A[i] == key:
            return i
    return -1

# Binary search
def binary_search(A, key, low, high):
    if low > high:
        return -1
    middle = (low + high) // 2
    if A[middle] == key:
        return middle
    elif A[middle] > key:
        return binary_search(A, key, low, middle - 1)
    else:
        return binary_search(A, key, middle + 1, high)

# Interpolation search
def binary_interpolation_search(A, key, low, high):
    if low > high or A[low] == A[high]:
        return -1
    middle = low + int((key - A[low]) * (high - low) / (A[high] - A[low]))
    if middle < low or middle > high:
        return -1
    if A[middle] == key:
        return middle
    elif A[middle] > key:
        return binary_interpolation_search(A, key, low, middle - 1)
    else:
        return binary_interpolation_search(A, key, middle + 1, high)

# Execution time collection
bubble_sort_times = []
insertion_sort_times = []
selection_sort_times = []
sequential_search_times = []
binary_search_times = []
interpolation_search_times = []

for _ in range(1000):  # Repeat 1000 times
    data = np.random.rand(100) * 100  # Data size of 100
    data1, data2, data3 = data.copy(), data.copy(), data.copy()

    # Sort times
    start = time.time()
    bubble_sort(data1)
    bubble_sort_times.append(time.time() - start)

    start = time.time()
    insertion_sort(data2)
    insertion_sort_times.append(time.time() - start)

    start = time.time()
    selection_sort(data3)
    selection_sort_times.append(time.time() - start)

    # Search times (use sorted data for binary and interpolation search)
    data1.sort()  # Binary and interpolation search require sorted data
    key = np.random.choice(data1)  # Choose a random key to search

    start = time.time()
    sequential_search(data1, key, 0, len(data1) - 1)
    sequential_search_times.append(time.time() - start)

    start = time.time()
    binary_search(data1, key, 0, len(data1) - 1)
    binary_search_times.append(time.time() - start)

    start = time.time()
    binary_interpolation_search(data1, key, 0, len(data1) - 1)
    interpolation_search_times.append(time.time() - start)

# Visualization
plt.figure(figsize=(20, 10))

# Subplot 1: Sorting Bar Plot
plt.subplot(2, 3, 1)
sorting_labels = ["Bubble Sort", "Insertion Sort", "Selection Sort"]
sorting_avg_times = [
    np.mean(bubble_sort_times),
    np.mean(insertion_sort_times),
    np.mean(selection_sort_times),
]
plt.bar(sorting_labels, sorting_avg_times, color=['blue', 'green', 'red'])
plt.title("Sorting Algorithms: Average Time (100 elements, 1000 iterations)")
plt.ylabel("Time (s)")

# Subplot 2: Search Bar Plot
plt.subplot(2, 3, 2)
search_labels = ["Sequential Search", "Binary Search", "Interpolation Search"]
search_avg_times = [
    np.mean(sequential_search_times),
    np.mean(binary_search_times),
    np.mean(interpolation_search_times),
]
plt.bar(search_labels, search_avg_times, color=['purple', 'orange', 'cyan'])
plt.title("Search Algorithms: Average Time (100 elements, 1000 iterations)")
plt.ylabel("Time (s)")

# Subplot 3: Sorting Box Plot
plt.subplot(2, 3, 3)
plt.boxplot([bubble_sort_times, insertion_sort_times, selection_sort_times],
            labels=sorting_labels)
plt.title("Sorting Algorithms: Time Distribution")
plt.ylabel("Time (s)")

# Subplot 4: Search Box Plot
plt.subplot(2, 3, 4)
plt.boxplot([sequential_search_times, binary_search_times, interpolation_search_times],
            labels=search_labels)
plt.title("Search Algorithms: Time Distribution")
plt.ylabel("Time (s)")

# Subplot 5: Sorting Scatter Plot
plt.subplot(2, 3, 5)
plt.scatter(range(len(bubble_sort_times)), bubble_sort_times, label="Bubble Sort", color='blue', alpha=0.7)
plt.scatter(range(len(insertion_sort_times)), insertion_sort_times, label="Insertion Sort", color='green', alpha=0.7)
plt.scatter(range(len(selection_sort_times)), selection_sort_times, label="Selection Sort", color='red', alpha=0.7)
plt.title("Sorting Algorithms: Scatter Plot")
plt.xlabel("Iteration")
plt.ylabel("Time (s)")
plt.legend()

# Subplot 6: Search Scatter Plot
plt.subplot(2, 3, 6)
plt.scatter(range(len(sequential_search_times)), sequential_search_times, label="Sequential Search", color='purple', alpha=0.7)
plt.scatter(range(len(binary_search_times)), binary_search_times, label="Binary Search", color='orange', alpha=0.7)
plt.scatter(range(len(interpolation_search_times)), interpolation_search_times, label="Interpolation Search", color='cyan', alpha=0.7)
plt.title("Search Algorithms: Scatter Plot")
plt.xlabel("Iteration")
plt.ylabel("Time (s)")
plt.legend()

plt.tight_layout()
plt.show()
```
